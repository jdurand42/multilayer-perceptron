```python
import numpy as np
import pandas as pd
```

```python
data_path = "../data/data.csv"

header = ['id', 'diagnosis']

for i in range(30):
    header.append(f'F{i}')

df = pd.read_csv(data_path, names=header)
df.head(5)

```

```python
df.describe()
```

```python
Y = df[['diagnosis']]

X = df.drop(labels=['diagnosis', 'id'], axis=1, inplace=False)
Y.head()
```

```python
X.head()
```

```python
Y = Y['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
Y.describe()
```

```python
Y
```

```python
# Normalization
# zscore

stds = []
means = []
# X = X.to_numpy()
for feature in X.columns:
    stds.append(X[feature].to_numpy().std())
    means.append(X[feature].to_numpy().mean())
print("stds: ", stds)
print("means: ", means)

raw = X.copy()

i = 0
for feature in X.columns:
    X[feature] = (X[feature] - means[i]) / stds[i]
    i += 1
```

```python
# print(raw.head(2))
X.head(2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python

import importlib
imported_module = importlib.import_module("MLP")
importlib.reload(imported_module)
from MLP import MultiLayerPerceptron

mlp = MultiLayerPerceptron(seed=1)

mlp.add_layer(size=4)
mlp.add_layer(size=8, label="hidden_layer")
mlp.add_layer(size=5, label="hidden_layer")
# mlp.add_layer(size=8, label="hidden_layer")
mlp.add_layer(label="output_layer")
# print(X.to_numpy())
mlp.fit(X_train.to_numpy(), Y_train.to_numpy(), verbose=False, epochs=1000)
```

```python
# evaluate

from sklearn.metrics import r2_score, accuracy_score
from sklearn.metrics import confusion_matrix
# import matplotlib

y_pred = mlp.predict(X_test.to_numpy())
print('score: ', accuracy_score(Y_test, y_pred))

cm = confusion_matrix(Y_test, y_pred)
print(cm)
```
